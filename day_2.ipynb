```python
#add and remove the data to the list

x= ['abhi', 'ram', 'shyam']
x.append('abhi')# add values at the end of the list
x.insert(0, 'abhi')# add values at the start of the list    
x.extend(['abhi', 'ram'])# add multiple values at the end of the list
x.remove('abhi')# remove the first occurence of the value
x.pop()# remove the last value of the list
x.clear()# remove all the values of the list
x = [1, 2, 3, 4, 5]
x.reverse()# reverse the list
x.sort()# sort the list in ascending order
x.sort(reverse=True)# sort the list in descending order
x = [1, 2, 3, 4, 5]
x = x[::-1]# reverse the list using slicing
x1 = x.copy()# create a copy of the list
```
